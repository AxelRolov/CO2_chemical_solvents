# Importing libraries

```python
import sys
from pathlib import Path
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse, r2_score as r2
from sklearn.model_selection import RepeatedKFold
from sklearn.ensemble import RandomForestRegressor
from utils import fragment_control, bounding_box, mkdir
```

# Predicted computed rate of reaction (R<sub>QSPR</sub>) vs computed rate of reaction (R<sub>MD</sub>) 

### Loading the dataframes

```python
dg_stats = pd.read_csv('../results/model_stats/dG_5x5CV_stats.csv')
cr_stats = pd.read_csv('../results/model_stats/CompRate_5x5CV_stats.csv')
```

### Calculating RMSE in CV

```python
rmse_cv_dg = mse(dg_stats[f'dG_MD_ylog'], dg_stats[f'avg_pred_fin'])**0.5
rmse_cv_cr = mse(cr_stats[f'CompRate_ylog'], cr_stats[f'avg_pred_fin'])**0.5
```

### Figure

```python
fig, axes = plt.subplots(1, 2, figsize=(20,10))

for axis in axes.reshape(1, -1)[0]:
    for tick in axis.xaxis.get_major_ticks():
        tick.label.set_fontsize(24) 
    for tick in axis.yaxis.get_major_ticks():
        tick.label.set_fontsize(24)
    for axisi in ['bottom','left']:
        axis.spines[axisi].set_linewidth(2.5)
    axis.spines['right'].set_visible(False)
    axis.spines['top'].set_visible(False)
    

# Plot for dG

gas_col = 'dG_MD_ylog'

axes[0].set_ylabel(r'$-lg(\Delta G_{QSPR})$', fontsize=34)
axes[0].set_xlabel(r'$-lg(\Delta G_{MD})$', fontsize=34)

axes[0].set_ylim(dg_stats[f'{gas_col}'].min()-0.01,  dg_stats[f'{gas_col}'].max()+0.01)
axes[0].set_xlim(dg_stats[f'{gas_col}'].min()-0.01,  dg_stats[f'{gas_col}'].max()+0.01)

rmse_cv = rmse_cv_dg
axes[0].plot([-1.9, dg_stats[f'{gas_col}'].max()], [(-1.9+2*rmse_cv), (dg_stats[f'{gas_col}'].max()+2*rmse_cv)], '--', c='black', linewidth=2.5)
axes[0].plot([-1.9, dg_stats[f'{gas_col}'].max()], [(-1.9-2*rmse_cv), (dg_stats[f'{gas_col}'].max()-2*rmse_cv)], '--', c='black', linewidth=2.5)
axes[0].plot([-1.9, dg_stats[f'{gas_col}'].max()], [-1.9, dg_stats[f'{gas_col}'].max()+0.001], '-', c='black', linewidth=2.5)

axes[0].scatter(dg_stats[f'{gas_col}'], dg_stats[f'avg_pred_fin'], s=154, c='black')
axes[0].scatter(dg_stats.loc[dg_stats['outlier']==1, f'{gas_col}'], dg_stats.loc[dg_stats['outlier']==1, f'avg_pred_fin'], c='red', s=154)
axes[0].scatter(dg_stats.loc[dg_stats['cid']=='105-59-9', f'{gas_col}'],
                   dg_stats.loc[dg_stats['cid']=='105-59-9', 'avg_pred_fin'],
                   s=304, c='springgreen', edgecolor='black', linewidth=1.5)


# Plot for the computed rate

gas_col = 'CompRate_ylog'
rmse_cv = rmse_cv_cr
axes[1].set_xlabel(r'$-lg(R_{MD})$', fontsize=34)
axes[1].set_ylabel(r'$-lg(R_{QSPR})$', fontsize=34)


axes[1].set_ylim(cr_stats[f'{gas_col}'].min()-0.05,  cr_stats[f'{gas_col}'].max()+0.05)
axes[1].set_xlim(cr_stats[f'{gas_col}'].min()-0.05,  cr_stats[f'{gas_col}'].max()+0.05)


axes[1].plot([-1.9, cr_stats[f'{gas_col}'].max()], [(-1.9+2*rmse_cv), (cr_stats[f'{gas_col}'].max()+2*rmse_cv)], '--', c='black', linewidth=2.5)
axes[1].plot([-1.9, cr_stats[f'{gas_col}'].max()], [(-1.9-2*rmse_cv), (cr_stats[f'{gas_col}'].max()-2*rmse_cv)], '--', c='black', linewidth=2.5)
axes[1].plot([-1.9, cr_stats[f'{gas_col}'].max()], [-1.9, cr_stats[f'{gas_col}'].max()+0.001], '-', c='black', linewidth=2.5)


axes[1].scatter(cr_stats[f'{gas_col}'], cr_stats[f'avg_pred_fin'], s=154, c='black')#, 'o', c='green')
axes[1].scatter(cr_stats.loc[cr_stats['outlier']==1, f'{gas_col}'], cr_stats.loc[cr_stats['outlier']==1, f'avg_pred_fin'], c='red', s=154)
axes[1].scatter(cr_stats.loc[cr_stats['cid']=='105-59-9', f'{gas_col}'],
                   cr_stats.loc[cr_stats['cid']=='105-59-9', 'avg_pred_fin'],
                   s=304, c='springgreen', edgecolor='black', linewidth=1.5)
```

# Figure for screening results

```python
screen_pred_cr = pd.read_csv('../results/screening/Comprate_pred.csv')  # computed rate predictions
screen_pred_dg = pd.read_csv('../results/screening/dG_pred.csv')  # dG predictions
```

```python
cols = ['std_smiles', 'source', 'RMD', 'GMD', 'pKa_b_pred', 'Slope_CO2', 'time_50CO2', 'cas']
screen_pred = pd.concat([screen_pred_cr.loc[:, cols+['comprate_antilog_avg_pred']], 
                         screen_pred_dg.loc[:, ['dg_antilog_avg_pred']]], axis=1) # combined dataset
```

```python
screen_pred['dg_antilog_avg_pred'] = screen_pred['dg_antilog_avg_pred'] * (-1)
```

```python
fig, axes = plt.subplots(2, 3, figsize=(30,20))

for axis in axes.reshape(1, -1)[0]:
    for tick in axis.xaxis.get_major_ticks():
        tick.label.set_fontsize(24) 
    for tick in axis.yaxis.get_major_ticks():
        tick.label.set_fontsize(24)
    for axisi in ['bottom','left']:
        axis.spines[axisi].set_linewidth(2.5)
    axis.spines['right'].set_visible(False)
    axis.spines['top'].set_visible(False)

# conditions for selecting relevant subsets

screened_cmpnds = (screen_pred['source']=='screening')&(screen_pred['pKa_b_pred']>=6)
md_cmpnds = (screen_pred['source']=='100MDamines')&(screen_pred['pKa_b_pred']>=6)
chowdhury_cmpnds = (screen_pred['source']=='Chowdhury')&(screen_pred['pKa_b_pred']>=6)
mdea = (screen_pred['cas']=='105-59-9')&(screen_pred['pKa_b_pred']>=6)
exp_cmpnds = (screen_pred_cr['source']=='screening')&(~screen_pred_cr['Slope_CO2'].isna())&(screen_pred['pKa_b_pred']>=6)

# predicted rate vs predicted pKa

axes[0, 0].set_ylabel(r'$R_{QSPR}$ (gL$^{-1}$min$^{-1}$)', fontsize=34)
axes[0, 0].set_xlabel(r'$pKa$', fontsize=34)
axes[0, 0].set_xlim(6., screen_pred_cr['pKa_b_pred'].max()+1)
axes[0, 0].set_ylim(screen_pred_cr['comprate_antilog_avg_pred'].min()-0.1,
                    screen_pred_cr['comprate_antilog_avg_pred'].max()+0.6)
axes[0, 0].scatter(screen_pred_cr.loc[screened_cmpnds, 'pKa_b_pred'], 
                   screen_pred_cr.loc[screened_cmpnds, 'comprate_antilog_avg_pred'],
                   s=154, c='grey', alpha=0.9)
axes[0, 0].scatter(screen_pred_cr.loc[md_cmpnds, 'pKa_b_pred'],
                   screen_pred_cr.loc[md_cmpnds, 'comprate_antilog_avg_pred'],
                   s=154, c='black')
axes[0, 0].scatter(screen_pred.loc[chowdhury_cmpnds, 'pKa_b_pred'],
                   screen_pred.loc[chowdhury_cmpnds, 'comprate_antilog_avg_pred'],
                   s=154, c='orange')
axes[0, 0].scatter(screen_pred.loc[mdea, 'pKa_b_pred'],
                   screen_pred.loc[mdea, 'comprate_antilog_avg_pred'],
                   s=304, c='springgreen', edgecolor='black', linewidth=1.5)
axes[0, 0].scatter(screen_pred.loc[exp_cmpnds,
                                      'pKa_b_pred'],
                   screen_pred.loc[exp_cmpnds,
                                    'comprate_antilog_avg_pred'],
                   s=154, c='blue')

# predicted rate vs deltaCO2

axes[0, 1].set_xlabel(r'$\Delta CO_{2}$ (s$^{-1}$)', fontsize=34)
axes[0, 1].set_ylabel(r'$R_{QSPR}$ (gL$^{-1}$min$^{-1}$)', fontsize=34)
axes[0, 1].set_xlim(screen_pred['Slope_CO2'].min()-0.0001, screen_pred['Slope_CO2'].max()+0.0005)
axes[0, 1].set_ylim(screen_pred['comprate_antilog_avg_pred'].min()-0.1,
                    screen_pred['comprate_antilog_avg_pred'].max()+0.6)
axes[0, 1].scatter(screen_pred.loc[chowdhury_cmpnds, 'Slope_CO2'], 
                   screen_pred.loc[chowdhury_cmpnds, 'comprate_antilog_avg_pred'], s=154, c='orange')
axes[0, 1].scatter(screen_pred.loc[md_cmpnds, 'Slope_CO2'],
                   screen_pred.loc[md_cmpnds, 'comprate_antilog_avg_pred'],
                   s=154, c='black')
axes[0, 1].scatter(screen_pred.loc[mdea, 'Slope_CO2'],
                   screen_pred.loc[mdea, 'comprate_antilog_avg_pred'],
                   s=304, c='springgreen', edgecolor='black', linewidth=1.5)
axes[0, 1].scatter(screen_pred.loc[exp_cmpnds, 'Slope_CO2'],
                   screen_pred.loc[exp_cmpnds, 'comprate_antilog_avg_pred'],
                   s=154, c='blue')

# deltaCO2 vs predicted pKa

axes[0, 2].set_ylabel(r'$\Delta CO_{2}$ (s$^{-1}$)', fontsize=34)
axes[0, 2].set_xlabel(r'$pKa$', fontsize=34)
axes[0, 2].set_ylim(screen_pred['Slope_CO2'].min()-0.0001, screen_pred['Slope_CO2'].max()+0.0005)
axes[0, 2].set_xlim(6., screen_pred['pKa_b_pred'].max()+1)
axes[0, 2].scatter(screen_pred.loc[chowdhury_cmpnds, 'pKa_b_pred'],
                   screen_pred.loc[chowdhury_cmpnds, 'Slope_CO2'],
                   s=154, c='orange')
axes[0, 2].scatter(screen_pred.loc[md_cmpnds, 'pKa_b_pred'],
                   screen_pred.loc[md_cmpnds, 'Slope_CO2'], 
                   s=154, c='black')
axes[0, 2].scatter(screen_pred.loc[mdea, 'pKa_b_pred'],
                   screen_pred.loc[mdea, 'Slope_CO2'],
                   s=304, c='springgreen', edgecolor='black', linewidth=1.5)
axes[0, 2].scatter(screen_pred.loc[exp_cmpnds, 'pKa_b_pred'],
                   screen_pred.loc[exp_cmpnds, 'Slope_CO2'], s=154, c='blue')


# dG vs predicted pKa

axes[1, 0].set_ylabel(r'$\Delta G_{QSPR}$ (kJmol$^{-1}$)', fontsize=34)
axes[1, 0].set_xlabel(r'$pK_{a}$', fontsize=34)
axes[1, 0].set_xlim(6., screen_pred['pKa_b_pred'].max()+1)
axes[1, 0].set_ylim(screen_pred['dg_antilog_avg_pred'].min()-5,
                    screen_pred['dg_antilog_avg_pred'].max()+5)
axes[1, 0].scatter(screen_pred.loc[screened_cmpnds, 'pKa_b_pred'],
                   screen_pred.loc[screened_cmpnds, 'dg_antilog_avg_pred'],
                   s=154, c='grey', alpha=0.9)
axes[1, 0].scatter(screen_pred.loc[md_cmpnds, 'pKa_b_pred'],
                   screen_pred.loc[md_cmpnds, 'dg_antilog_avg_pred'],
                   s=154, c='black')
axes[1, 0].scatter(screen_pred.loc[chowdhury_cmpnds, 'pKa_b_pred'],
                   screen_pred.loc[chowdhury_cmpnds, 'dg_antilog_avg_pred'],
                   s=154, c='orange')
axes[1, 0].scatter(screen_pred.loc[mdea, 'pKa_b_pred'],
                   screen_pred.loc[mdea, 'dg_antilog_avg_pred'],
                   s=304, c='springgreen', edgecolor='black', linewidth=1.5)
axes[1, 0].scatter(screen_pred.loc[exp_cmpnds, 'pKa_b_pred'],
                   screen_pred.loc[exp_cmpnds, 'dg_antilog_avg_pred'],
                   s=154, c='blue')

# dG vs deltaCO2

axes[1, 1].set_xlabel(r'$\Delta CO_{2}$ (s$^{-1}$)', fontsize=34)
axes[1, 1].set_ylabel(r'$\Delta G_{QSPR}$ (kJmol$^{-1}$)', fontsize=34)
axes[1, 1].set_xlim(screen_pred['Slope_CO2'].min()-0.0001, screen_pred['Slope_CO2'].max()+0.0005)
axes[1, 1].set_ylim(screen_pred['dg_antilog_avg_pred'].min()-5,
                    screen_pred['dg_antilog_avg_pred'].max()+5)
axes[1, 1].scatter(screen_pred.loc[chowdhury_cmpnds, 'Slope_CO2'],
                   screen_pred.loc[chowdhury_cmpnds, 'dg_antilog_avg_pred'],
                   s=154, c='orange')
axes[1, 1].scatter(screen_pred.loc[md_cmpnds, 'Slope_CO2'], 
                   screen_pred.loc[md_cmpnds, 'dg_antilog_avg_pred'],
                   s=154, c='black')
axes[1, 1].scatter(screen_pred.loc[mdea, 'Slope_CO2'],
                   screen_pred.loc[mdea, 'dg_antilog_avg_pred'],
                   s=304, c='springgreen', edgecolor='black', linewidth=1.5)
axes[1, 1].scatter(screen_pred.loc[exp_cmpnds, 'Slope_CO2'],
                   screen_pred.loc[exp_cmpnds, 'dg_antilog_avg_pred'],
                   s=154, c='blue')

# dG vs predicted rate

axes[1, 2].set_ylabel(r'$\Delta G_{QSPR}$ (kJmol$^{-1}$)', fontsize=34)
axes[1, 2].set_xlabel(r'$R_{QSPR}$ (gL$^{-1}$min$^{-1}$)', fontsize=34)
axes[1, 2].set_xlim(screen_pred['comprate_antilog_avg_pred'].min()-0.6,
                    screen_pred['comprate_antilog_avg_pred'].max()+1)
axes[1, 2].set_ylim(screen_pred['dg_antilog_avg_pred'].min()-5, 
                    screen_pred['dg_antilog_avg_pred'].max()+5)
axes[1, 2].scatter(screen_pred.loc[screened_cmpnds, 'comprate_antilog_avg_pred'],
                   screen_pred.loc[screened_cmpnds, 'dg_antilog_avg_pred'],
                   s=154, c='grey', alpha=0.9)
axes[1, 2].scatter(screen_pred.loc[md_cmpnds, 'comprate_antilog_avg_pred'],
                   screen_pred.loc[md_cmpnds, 'dg_antilog_avg_pred'],
                   s=154, c='black')
axes[1, 2].scatter(screen_pred.loc[chowdhury_cmpnds, 'comprate_antilog_avg_pred'], 
                   screen_pred.loc[chowdhury_cmpnds, 'dg_antilog_avg_pred'],
                   s=154, c='orange')
axes[1, 2].scatter(screen_pred.loc[exp_cmpnds, 'comprate_antilog_avg_pred'],
                   screen_pred.loc[exp_cmpnds, 'dg_antilog_avg_pred'],
                   s=154, c='blue')
axes[1, 2].scatter(screen_pred.loc[mdea, 'comprate_antilog_avg_pred'],
                   screen_pred.loc[mdea, 'dg_antilog_avg_pred'],
                   s=304, c='springgreen', edgecolor='black', linewidth=1.5)
plt.show()
```
